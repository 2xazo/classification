```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("iris.csv")
```

```python
from sklearn.preprocessing import StandardScaler

# Extract the numerical data from the specified columns
# Replace 'SepalLengthcm', 'SepalWidthcm' with the actual column names

features_data = data[['SepalLengthCm', 'SepalWidthCm']].values # Corrected column names, assuming 'Cm' instead of 'cm'

# Now, apply StandardScaler to the numerical data
features_scaled = StandardScaler().fit_transform(features_data)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(features_scaled )
```

```python
# Plotting the clusters
plt.figure(figsize=(8, 4))
plt.scatter(data['SepalLengthCm'], data['SepalWidthCm'], c=clusters, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('K-means Clustering of Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.colorbar(label='Cluster Label')
plt.show()
```
